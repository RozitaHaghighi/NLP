**Balanced mini dataset with 3 categories (Technology, History, Biology)**

```python
def assign_label_from_title(title):
    title = title.lower()
    if any(kw in title for kw in ['computer', 'artificial intelligence', 'software', 'technology', 'internet']):
        return 0  # Technology
    elif any(kw in title for kw in ['revolution', 'war', 'empire', 'president', 'history']):
        return 1  # History/Politics
    elif any(kw in title for kw in ['biology', 'photosynthesis', 'organism', 'cell', 'genetics']):
        return 2  # Biology
    else:
        return 3  # Unknown / Skip

# Only keep long articles and label those with known categories
labeled_data = []

for item in wiki:
    if len(item['text'].split()) < 100:
        continue
    label = assign_label_from_title(item['title'])
    if label != 3:
        labeled_data.append({
            'title': item['title'],
            'text': item['text'],
            'label': label
        })
    if len(labeled_data) >= 6:
        break  # Get just 6 for now

# Show preview
print(f"\nTotal labeled articles: {len(labeled_data)}")
print("Sample title:", labeled_data[0]['title'])
print("Label:", labeled_data[0]['label'])
print("Text preview:\n", labeled_data[0]['text'][:500])
```

**Hierarchical Tokenization (Sentence → Words)**

**A list of sentences**

**Where each sentence is a list of tokens (words)**

This prepares the data for the HAN model (which processes inputs as Document → Sentences → Words).

```python
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize

nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('all', halt_on_error=False)


def tokenize_hierarchical(text):
    sentences = sent_tokenize(text)
    return [word_tokenize(sent.lower()) for sent in sentences]

# Apply to each article
for item in labeled_data:
    item['tokens'] = tokenize_hierarchical(item['text'])

# Preview one tokenized article
print("Title:", labeled_data[0]['title'])
print("Label:", labeled_data[0]['label'])
print("First 2 sentences (tokenized):", labeled_data[0]['tokens'][:2])
```

**Build Vocabulary + Load fastText Embeddings**

Build a vocabulary from all the words in my tokenized dataset

Load pre-trained fastText vectors (like cc.en.300.vec.gz)

Create an embedding matrix matching my vocab

```python
import os

url = "https://dl.fbaipublicfiles.com/fasttext/vectors-crawl/cc.en.300.vec.gz"
filename = "cc.en.300.vec.gz"

if not os.path.exists(filename):
    print("Downloading fastText embeddings (~6GB)...")
    os.system(f"wget {url}")
else:
    print("File already exists.")
```

```python
from gensim.models import KeyedVectors
from collections import Counter
import numpy as np

# Load fastText vectors (this may take 2–5 mins)
print("Loading fastText embeddings...")
ft_model = KeyedVectors.load_word2vec_format("cc.en.300.vec.gz", binary=False)
print("fastText loaded.")

# Build vocab from tokenized articles
word_counter = Counter()
for item in labeled_data:
    for sentence in item['tokens']:
        word_counter.update(sentence)

# Keep top N words (optional for limiting vocab size)
vocab_limit = 10000
most_common_words = [word for word, _ in word_counter.most_common(vocab_limit)]

# Assign word → index
word2idx = {"<PAD>": 0, "<UNK>": 1}
for word in most_common_words:
    if word in ft_model:
        word2idx[word] = len(word2idx)

# Create embedding matrix
embedding_dim = 300
embedding_matrix = np.zeros((len(word2idx), embedding_dim))

for word, idx in word2idx.items():
    if word in ft_model:
        embedding_matrix[idx] = ft_model[word]
    else:
        embedding_matrix[idx] = np.random.normal(scale=0.6, size=(embedding_dim,))

print(f"Vocabulary size: {len(word2idx)}")
```

**Encode Documents into Tensors**

```python
import torch

# Set fixed limits (you can increase later)
MAX_SENTENCES = 10
MAX_WORDS = 20

def encode_article(tokens, word2idx, max_sents=MAX_SENTENCES, max_words=MAX_WORDS):
    encoded_doc = []

    for sent in tokens[:max_sents]:
        encoded_sent = [word2idx.get(word, word2idx["<UNK>"]) for word in sent[:max_words]]
        # Pad sentence if too short
        while len(encoded_sent) < max_words:
            encoded_sent.append(word2idx["<PAD>"])
        encoded_doc.append(encoded_sent)

    # Pad doc if too few sentences
    while len(encoded_doc) < max_sents:
        encoded_doc.append([word2idx["<PAD>"]] * max_words)

    return encoded_doc

# Encode all articles
X_data = [encode_article(item['tokens'], word2idx) for item in labeled_data]
y_data = [item['label'] for item in labeled_data]

# Convert to PyTorch tensors
X_tensor = torch.tensor(X_data, dtype=torch.long)
y_tensor = torch.tensor(y_data, dtype=torch.long)

print("X_tensor shape:", X_tensor.shape)  # (batch_size, max_sentences, max_words)
print("y_tensor shape:", y_tensor.shape)  # (batch_size,)
```

Build the HAN Model in PyTorch

Word-level encoder: Bi-GRU + word attention

Sentence-level encoder: Bi-GRU + sentence attention

Output: classification layer

```python
import torch.nn as nn
import torch.nn.functional as F

class Attention(nn.Module):
    def __init__(self, hidden_dim):
        super().__init__()
        self.attn = nn.Linear(hidden_dim, 1)

    def forward(self, x):
        weights = self.attn(x).squeeze(-1)  # (batch, seq_len)
        weights = F.softmax(weights, dim=1)
        weighted = (x * weights.unsqueeze(-1)).sum(dim=1)  # (batch, hidden_dim)
        return weighted, weights  # return both for attention tracking

class HAN(nn.Module):
    def __init__(self, vocab_size, embedding_dim, hidden_dim, num_classes, embedding_matrix):
        super().__init__()
        self.embedding = nn.Embedding(vocab_size, embedding_dim, padding_idx=0)
        self.embedding.weight.data.copy_(torch.from_numpy(embedding_matrix))
        self.embedding.weight.requires_grad = False

        self.word_gru = nn.GRU(embedding_dim, hidden_dim, bidirectional=True, batch_first=True)
        self.word_attn = Attention(hidden_dim * 2)

        self.sent_gru = nn.GRU(hidden_dim * 2, hidden_dim, bidirectional=True, batch_first=True)
        self.sent_attn = Attention(hidden_dim * 2)

        self.fc = nn.Linear(hidden_dim * 2, num_classes)

    def forward(self, x, return_attn=False):
        B, S, W = x.size()
        x = x.view(B * S, W)
        word_embed = self.embedding(x)
        word_out, _ = self.word_gru(word_embed)
        sent_vec, word_attn_weights = self.word_attn(word_out)
        sent_vec = sent_vec.view(B, S, -1)
        sent_out, _ = self.sent_gru(sent_vec)
        doc_vec, sent_attn_weights = self.sent_attn(sent_out)
        output = self.fc(doc_vec)
        if return_attn:
            return output, word_attn_weights.view(B, S, -1), sent_attn_weights
        return output
```

```python
from torch.utils.data import TensorDataset, DataLoader

# Training config
batch_size = 2
epochs = 5
num_classes = 3
hidden_dim = 50  # You can tune this

# DataLoader
dataset = TensorDataset(X_tensor, y_tensor)
dataloader = DataLoader(dataset, batch_size=batch_size, shuffle=True)

# Model, optimizer, loss
model = HAN(
    vocab_size=len(word2idx),
    embedding_dim=300,
    hidden_dim=50,
    num_classes=3,
    embedding_matrix=embedding_matrix
)

optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
criterion = nn.CrossEntropyLoss()

# Training loop
model.train()
for epoch in range(epochs):
    total_loss = 0
    correct = 0
    total = 0

    for batch_x, batch_y in dataloader:
        optimizer.zero_grad()
        outputs = model(batch_x)
        loss = criterion(outputs, batch_y)
        loss.backward()
        optimizer.step()

        total_loss += loss.item()
        preds = torch.argmax(outputs, dim=1)
        correct += (preds == batch_y).sum().item()
        total += batch_y.size(0)

    acc = correct / total
    print(f"Epoch {epoch+1}/{epochs} - Loss: {total_loss:.4f} - Accuracy: {acc:.2f}")
```

```python
import numpy as np

# Choose an article to interpret
index = 0
sample_x = X_tensor[index].unsqueeze(0)  # Shape: (1, S, W)
sample_tokens = labeled_data[index]['tokens']
title = labeled_data[index]['title']

# Run model with attention output
model.eval()
with torch.no_grad():
    output, word_attns, sent_attns = model(sample_x, return_attn=True)

# Convert attention weights to numpy
word_attns = word_attns.squeeze(0).cpu().numpy()  # (S, W)
sent_attns = sent_attns.squeeze(0).cpu().numpy()  # (S,)

# Visualize
print(f"\n📝 Title: {title}")
print("\n🔍 Sentence-Level Attention:\n")

for i, sentence in enumerate(sample_tokens):
    if i >= sent_attns.shape[0]: break
    score = sent_attns[i]
    sentence_text = " ".join(sentence)
    print(f"\n📌 Sentence {i+1} [Attention: {score:.3f}]:\n{sentence_text}")

    word_scores = word_attns[i][:len(sentence)]
    top_word_ids = np.argsort(word_scores)[::-1][:3]
    top_words = [sentence[j] for j in top_word_ids]
    print("⭐ Top words:", ", ".join(top_words))
```

```python
import matplotlib.pyplot as plt

def plot_attention_heatmap(sent_attns, word_attns, tokens, title="Attention Heatmap"):
    import matplotlib.pyplot as plt
    import numpy as np

    num_sents = len(tokens)
    real_sent_attns = sent_attns[:num_sents]

    fig, axs = plt.subplots(1, 2, figsize=(16, 6))

    # Sentence-level attention bar chart
    axs[0].bar(range(num_sents), real_sent_attns)
    axs[0].set_title("Sentence-Level Attention")
    axs[0].set_xlabel("Sentence Index")
    axs[0].set_ylabel("Attention Weight")

    # Word-level heatmap for top 3 sentences
    top_sents = np.argsort(real_sent_attns)[::-1][:3]
    word_heat = np.array([word_attns[i][:len(tokens[i])] for i in top_sents], dtype=object)

    # Create padded word heatmap
    max_words = max(len(row) for row in word_heat)
    padded_heat = np.array([np.pad(row, (0, max_words - len(row)), constant_values=np.nan) for row in word_heat])

    im = axs[1].imshow(padded_heat, cmap='viridis', aspect='auto')
    axs[1].set_title("Word-Level Attention (Top 3 Sentences)")
    axs[1].set_xlabel("Word Index")
    axs[1].set_yticks(range(len(top_sents)))
    axs[1].set_yticklabels([f"Sentence {i+1}" for i in top_sents])
    plt.colorbar(im, ax=axs[1])

    plt.suptitle(title)
    plt.tight_layout()
    plt.show()
plot_attention_heatmap(sent_attns, word_attns, sample_tokens, title=title)

```

```python
from IPython.core.display import display, HTML
import matplotlib

def get_color(score, cmap='YlOrRd'):
    import matplotlib
    color = matplotlib.cm.get_cmap(cmap)(score)  # RGBA
    return f'rgb({int(color[0]*255)}, {int(color[1]*255)}, {int(color[2]*255)})'

def render_colored_attention(tokens, word_attns, sent_attns, top_k_sentences=3):
    # Limit to actual sentence count
    num_real_sentences = len(tokens)
    real_sent_attns = sent_attns[:num_real_sentences]
    top_sent_ids = np.argsort(real_sent_attns)[::-1][:top_k_sentences]

    html = "<h3>📝 Attention Highlight</h3>"
    for i in top_sent_ids:
        if i >= len(tokens): continue  # prevent IndexError
        sent_html = f"<div style='margin-bottom:10px;'><b>Sentence {i+1}:</b><br>"
        for j, word in enumerate(tokens[i]):
            if j >= word_attns.shape[-1]: continue
            color = get_color(word_attns[i][j])
            sent_html += f"<span style='background-color:{color}; padding:2px 4px; margin:1px; border-radius:4px'>{word}</span> "
        sent_html += "</div>"
        html += sent_html

    display(HTML(html))

render_colored_attention(sample_tokens, word_attns, sent_attns)
```
